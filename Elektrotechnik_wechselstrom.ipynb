```python
import tkinter as tk
import numpy as np


main = tk.Tk()
main.minsize(800,400)
main.title("Elektrotechnik")

labelframe = tk.LabelFrame(main)
buttonframe = tk.Frame(main)

#Label und Input

lbl_spannung = tk.Label(labelframe, text="Spannung (V): ", font=15)
lbl_spannung.grid(row=0,column=0)

txt_spannung = tk.Entry(labelframe, width=5,font=15)
txt_spannung.grid(row=0,column=1)

lbl_strom = tk.Label(labelframe, text="Strom (A): ", font=15)
lbl_strom.grid(row=0, column=2)

txt_strom = tk.Entry(labelframe, width=5,font=15)
txt_strom.grid(row=0,column=3)

lbl_widerstand = tk.Label(labelframe, text="Widerstand (R): ", font=15)
lbl_widerstand.grid(row=0,column=4)

txt_widerstand = tk.Entry(labelframe,width=5,font=15)
txt_widerstand.grid(row=0,column=5)

lbl_leistung = tk.Label(labelframe, text="Leistung (W): ",font=15)
lbl_leistung.grid(row=0,column=6)

txt_leistung = tk.Entry(labelframe,width=5,font=15)
txt_leistung.grid(row=0,column=7)

lbl_zeit = tk.Label(labelframe, text="Zeit (s): ",font=15)
lbl_zeit.grid(row=0,column=8)

txt_zeit = tk.Entry(labelframe, width=5,font=15)
txt_zeit.grid(row=0,column=9)

lbl_frequenz = tk.Label(labelframe, text="Frequenz (Hz):", font=15)
lbl_frequenz.grid(row=1,column=0)

txt_frequenz = tk.Entry(labelframe, width=5,font=15)
txt_frequenz.grid(row=1,column=1)

lbl_phasenverschiebung = tk.Label(labelframe, text="Phasenversch. (°):",font=15)
lbl_phasenverschiebung.grid(row=1,column=2)

txt_phasenverschiebung = tk.Entry(labelframe, width=5,font=15)
txt_phasenverschiebung.grid(row=1,column=3)

lbl_induktivität = tk.Label(labelframe,text="Induktivität (H):",font=15)
lbl_induktivität.grid(row=1,column=4)

txt_induktivität = tk.Entry(labelframe,width=5,font=15)
txt_induktivität.grid(row=1,column=5)

lbl_pole = tk.Label(labelframe,text="Pole (1,2,3): ", font=15)
lbl_pole.grid(row=1,column=6)

txt_pole =tk.Entry(labelframe,width=5,font=15)
txt_pole.grid(row=1,column=7)

lbl_kreisfrequenz = tk.Label(labelframe,text="Kreisfrequenz (1/s):",font=15)
lbl_kreisfrequenz.grid(row=1,column=8)

txt_kreisfrequenz = tk.Entry(labelframe,width=5,font=15)
txt_kreisfrequenz.grid(row=1,column=9)

lbl_ladungsmenge = tk.Label(labelframe,text="Ladungsmenge (C):",font=15)
lbl_ladungsmenge.grid(row=2,column=0)

txt_ladungsmenge = tk.Entry(labelframe,width=5,font=15)
txt_ladungsmenge.grid(row=2,column=1)

lbl_länge = tk.Label(labelframe, text="Länge (m):", font=15)
lbl_länge.grid(row=2,column=2)

txt_länge = tk.Entry(labelframe,width=5,font=15)
txt_länge.grid(row=2,column=3)

lbl_querschnitt = tk.Label(labelframe,text="Querschnitt (mm2):",font=15)
lbl_querschnitt.grid(row=2,column=4)

txt_querschnitt = tk.Entry(labelframe,width=5,font=15)
txt_querschnitt.grid(row=2,column=5)


# FunctionsFunctions

def berechne_spannung():
    strom_str = txt_strom.get()
    widerstand_str = txt_widerstand.get()
    leistung_str = txt_leistung.get()

    if strom_str and widerstand_str:
        out = float(strom_str) * float(widerstand_str)
        output.config(text=f"Spannugn ist {round(out,3)} V \n I * R")

    elif leistung_str and strom_str:
        out1 = float(leistung_str) / float(strom_str)
        output.config(text=f"Spannugn ist {round(out1,3)} V \n W / I")

    elif leistung_str and widerstand_str:
        out2 = np.sqrt((float(leistung_str)*float(widerstand_str)))
        output.config(text=f"Spannung ist {round(out2,3)}V \n sqrt( P * R)")

    else:
        txt_strom.config(background="yellow")
        txt_widerstand.config(background="yellow")
        txt_leistung.config(background="yellow")

        txt_strom.after(2000, lambda: txt_strom.config(background="white"))
        txt_widerstand.after(2000, lambda: txt_widerstand.config(background="white"))
        txt_leistung.after(2000, lambda: txt_leistung.config(background="white"))

        output.config(text="2 vo dene Fäuder usfüue")

def berechne_strom():
    spannung = txt_spannung.get()
    widerstand = txt_widerstand.get()
    leistung = txt_leistung.get()
    if spannung and widerstand:
        out = float(spannung) / float(widerstand)
        output.config(text=f"Strom ist {round(out,3)} A \n U / R")
    elif leistung and spannung:
        out1 = float(leistung) / float(spannung)
        output.config(text=f"Storm ist {round(out1,3)} A \n P / U")
    elif leistung and widerstand:
        out2 = np.sqrt((float(leistung)/float(widerstand)))
        output.config(text=f"Strom ist {round(out2,3)} A \n sqrt( P / R )")
    else:
        txt_spannung.config(background="yellow")
        txt_widerstand.config(background="yellow")
        txt_leistung.config(background="yellow")

        txt_spannung.after(2000, lambda: txt_spannung.config(background="white"))
        txt_widerstand.after(2000, lambda: txt_widerstand.config(background="white"))
        txt_leistung.after(2000,lambda: txt_leistung.config(background="white"))

        output.config(text="bruche meh infos?! \n Spannung + Widerstand \n Leistung + Spannung \n Leistung + Widerstand")

def berechen_widerstand():
    spannung = txt_spannung.get()
    strom = txt_strom.get()
    leistung = txt_leistung.get()
    if spannung and strom:
        out = float(spannung) / float(strom)
        output.config(text=f"Widerstand ist {round(out,3)} Ohm \n U / I")
    elif leistung and strom:
        out1 = leistung / (strom**2)
        output.config(text=f"Widerstand ist {round(out1,3)} Ohm \n P / I**2")
    elif spannung and leistung:
        out3 = (float(spannung)**2) / float(leistung)
        output.config(text=f"Widerstand ist {round(out3,3)} Ohm \n U**2 / P")
    else:
        txt_spannung.config(background="yellow")
        txt_strom.config(background="yellow")
        txt_leistung.config(background="yellow")

        txt_spannung.after(2000, lambda: txt_spannung.config(background="white"))
        txt_strom.after(2000, lambda: txt_strom.config(background="white"))
        txt_leistung.after(2000, lambda: txt_leistung.config(background="white"))

        output.config(text="so chash nid rächne, oder emu nid guet ;)\n Spannugn + Strom \n Leistung + Strom \n Spannung + Leistung")

def berechne_leistung():
    spannung = txt_spannung.get()
    strom = txt_strom.get()
    widerstand = txt_widerstand.get()
    if spannung and strom:
        out = float(spannung) * float(strom)
        output.config(text=f"Leistung ist {round(out,3)} W \n U * I")
    elif spannung and widerstand:
        out1 = float(spannung)**2 / float(widerstand)
        output.config(text=f"Leistung ist {round(out1,3)} W \n U**2 / R")
    elif strom and widerstand:
        out2 = float(strom)**2 * float(widerstand)
        output.config(text=f"Leistung ist {round(out2,3)} W \n I**2 * R")
    else:
        txt_spannung.config(background="yellow")
        txt_strom.config(background="yellow")
        txt_widerstand.config(background="yellow")

        txt_spannung.after(2000, lambda: txt_spannung.config(background="white"))
        txt_strom.after(2000, lambda: txt_strom.config(background="white"))
        txt_widerstand.after(2000, lambda: txt_widerstand.config(background="white"))

        output.config(text="das chas iz aber nid si ?!\n Spannung + Strom\n Spannung + Widerstand \n Strom + Widerstand")

def berechne_leiterwiderstand():    # evtl Spezifischen Widerstand überprüfen
    länge = txt_länge.get()
    querschnitt = txt_querschnitt.get()
    
    
    if länge and querschnitt:
        leiterwiderstand = 0.01786 * (float(länge)/float(querschnitt))
        leitungsleitwert = 1/leiterwiderstand
        output.config(text=f"Leiterwiderstand ist {round(leiterwiderstand,5)} Ohm \n Leitungsleitwert ist {round(leitungsleitwert,3)} \nEinheitswiderstand von Kupfer (0,01786) * (Länge / Querschnitt)")
    else:
        txt_länge.config(background="yellow")
        txt_querschnitt.config(background="yellow")

        txt_länge.after(2000, lambda: txt_länge.config(background="white"))
        txt_querschnitt.after(2000, lambda: txt_querschnitt.config(background="white"))

        output.config(text="Brauche Länge und Querschnitt")

def berechen_r(txt_paralell, lbl_r_resultat):
    zahlen = txt_paralell.get()
    zahlen1 = [float(x) for x in zahlen.split()]
    result=0
    for i in zahlen1:
        result += 1/i
    resulttot = 1/result
    lbl_r_resultat.config(text=f"Gesamtwiderstand der Parallelschaltung {round(resulttot,3)} Ohm")

def öffne_r_rechner():
    r_rechner = tk.Toplevel(main)
    r_rechner.minsize(600,400)
    r_rechner.title("Widerstandsrechner")

    parallelframe = tk.LabelFrame(r_rechner)

    lbl_paralell = tk.Label(parallelframe, text="Parallel Widerstände eingeben:",font=15)
    lbl_paralell.pack(expand=True)

    txt_paralell = tk.Entry(parallelframe,width=25, font=15)
    txt_paralell.pack(expand=True)

    btn_r_result = tk.Button(parallelframe, text="Berechne Gesamtwiderstand",font=15, command=lambda: berechen_r(txt_paralell, lbl_r_resultat))
    btn_r_result.pack(expand=True)

    lbl_r_resultat = tk.Label(parallelframe, text="Output",font=15)
    lbl_r_resultat.pack(expand=True)

    parallelframe.pack()
    


#Buttonframe

btn_leistung = tk.Button(buttonframe, text="Leistung", font=15, command=berechne_leistung)
btn_leistung.grid(row=0,column=0)

btn_strom = tk.Button(buttonframe, text="Strom", font=15,command=berechne_strom)
btn_strom.grid(row=0,column=2)

btn_spannung = tk.Button(buttonframe, text="Spannung", font=15,command=berechne_spannung)
btn_spannung.grid(row=0,column=3)

btn_widerstand = tk.Button(buttonframe, text="Widerstand", font=15, command=berechen_widerstand)
btn_widerstand.grid(row=0,column=4)

btn_widerstandsrechenr = tk.Button(buttonframe, text="R Rechner", font=15, fg="blue", command=öffne_r_rechner)
btn_widerstandsrechenr.grid(row=0,column=5)

btn_leiterwiderstand = tk.Button(buttonframe,text="Leiterwiderstand",font=15,command=berechne_leiterwiderstand)
btn_leiterwiderstand.grid(row=0,column=6)



#Output

output = tk.Label(main, text="Output",font=15)




labelframe.pack()
buttonframe.pack()
output.pack(side="left",expand=True)


main.mainloop()
```
