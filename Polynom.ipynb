```python
import tkinter as tk
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

main = tk.Tk()
main.minsize(800,400)
main.title("Polynome")

def multiplikation():
    p1_input = txtpoly1.get()
    p2_input = txtpoly2.get()

    p1_coeffs = [float(x) for x in p1_input.split()]
    p2_coeffs = [float(x) for x in p2_input.split()]

    p1 = np.poly1d(p1_coeffs)
    p2 = np.poly1d(p2_coeffs)

    result = p1 * p2
    lbl_resultp1["text"] = str(result)

def division():
    p1_input = txtpoly1.get()
    p2_input = txtpoly2.get()

    p1_coeffs = [float(x) for x in p1_input.split()]
    p2_coeffs = [float(x) for x in p2_input.split()]

    p1 = np.poly1d(p1_coeffs)
    p2 = np.poly1d(p2_coeffs)

    q, r = np.polydiv(p1, p2) 

    lbl_resultp1["text"] = "Quotient:" + str(q) + "\n" + "Rest:" + str(r)
    

def nullstellen():
    p1_input = txtpoly1.get()
    p2_input = txtpoly2.get()

    p1_coeffs = [float(x) for x in p1_input.split()]
    p2_coeffs = [float(x) for x in p2_input.split()]

    p1 = np.poly1d(p1_coeffs)
    p2 = np.poly1d(p2_coeffs)

    p01=np.roots(p1)
    p02=np.roots(p2)

    lbl_resultp1["text"] = "Nullstellen = " + str(p01)
    lbl_resultp2["text"] = "Nullstellen = " + str(p02)


def gerade():
    p1_input = txtpoly1.get()
    p2_input = txtpoly2.get()

    p1_coeffs = [float(x) for x in p1_input.split()]
    p2_coeffs = [float(x) for x in p2_input.split()]

    p1 = np.poly1d(p1_coeffs)
    p2 = np.poly1d(p2_coeffs)

    o1=p1.order
    o2=p2.order

    lbl_resultp1["text"] = "Polynom 1 ist "+ str(o1) + ".Gerades"
    lbl_resultp2["text"] = "Polynom 2 ist "+ str(o2) + ".Gerades"

def funktionswert():
    p1_input = txtpoly1.get()
    p2_input = txtpoly2.get()

    x=int(txt_funktionswert.get())

    p1_coeffs = [float(x) for x in p1_input.split()]
    p2_coeffs = [float(x) for x in p2_input.split()]

    p1 = np.poly1d(p1_coeffs)
    p2 = np.poly1d(p2_coeffs)

    lbl_resultp1["text"] = "Funktionswert P1 :" + str(x) +"=    " + str((p1(x)))
    lbl_resultp2["text"] = "Funktionswert P2 :" + str(x) +"=    " + str((p2(x)))

def ableitung1():
    p1_input = txtpoly1.get()
    p2_input = txtpoly2.get()

    p1_coeffs = [float(x) for x in p1_input.split()]
    p2_coeffs = [float(x) for x in p2_input.split()]

    p1 = np.poly1d(p1_coeffs)
    p2 = np.poly1d(p2_coeffs)

    abl1 = p1.deriv(1)
    abl2=p2.deriv(1)

    lbl_resultp1["text"] = str(p1) + "\n1. Ableitung von P1:\n" + str(abl1)
    lbl_resultp2["text"] = str(p2) + "\n1. Ableitung von P2:\n" + str(abl2)

def ableitung2():
    p1_input = txtpoly1.get()
    p2_input = txtpoly2.get()

    p1_coeffs = [float(x) for x in p1_input.split()]
    p2_coeffs = [float(x) for x in p2_input.split()]

    p1 = np.poly1d(p1_coeffs)
    p2 = np.poly1d(p2_coeffs)

    abl1 = p1.deriv(2)
    abl2=p2.deriv(2)

    lbl_resultp1["text"] = str(p1) + "\n1. Ableitung von P1:\n" + str(abl1)
    lbl_resultp2["text"] = str(p2) + "\n1. Ableitung von P2:\n" + str(abl2)

def ableitung3():
    p1_input = txtpoly1.get()
    p2_input = txtpoly2.get()

    p1_coeffs = [float(x) for x in p1_input.split()]
    p2_coeffs = [float(x) for x in p2_input.split()]

    p1 = np.poly1d(p1_coeffs)
    p2 = np.poly1d(p2_coeffs)

    abl1 = p1.deriv(3)
    abl2=p2.deriv(3)

    lbl_resultp1["text"] = str(p1) + "\n1. Ableitung von P1:\n" + str(abl1)
    lbl_resultp2["text"] = str(p2) + "\n1. Ableitung von P2:\n" + str(abl2)

def stammfunktion():
    p1_input = txtpoly1.get()
    p2_input = txtpoly2.get()

    p1_coeffs = [float(x) for x in p1_input.split()]
    p2_coeffs = [float(x) for x in p2_input.split()]

    p1 = np.poly1d(p1_coeffs)
    p2 = np.poly1d(p2_coeffs)

    stmf1=p1.integ()
    stmf2=p2.integ()

    lbl_resultp1["text"] = "Stammfunktion P1:\n" + str(stmf1)
    lbl_resultp2["text"] = "Stammfunktion P2:\n" + str(stmf2)

def vietap():
    
    x1_input = txtx1.get()
    x2_input = txtx2.get()

    p = int(x1_input) + int(x2_input)

    lbl_resultp1["text"] = "Satz von Vieta p:\n" + "-"+ str(p)
    

def vietaq():

    x1_input = txtx1.get()
    x2_input = txtx2.get()

    q = int(x1_input)*int(x2_input)

    lbl_resultp1["text"] = "Satz von Vieta q:\n" + str(q)


def plot1():
    new_window = tk.Tk()
    new_window.title("Plot 1")

    fig, ax = plt.subplots()

    p1_input = txtpoly1.get()

    p1_coeffs = [float(x) for x in p1_input.split()]
    p1 = np.poly1d(p1_coeffs)

    x = np.linspace(-5, 5, 100)
    y = p1(x)
    ax.plot(x, y)
    ax.grid(True)

    poly_label = tk.Label(new_window, text=str(p1))
    poly_label.pack()

    canvas = FigureCanvasTkAgg(fig, master=new_window)
    canvas.draw()

    canvas.get_tk_widget().pack()

    new_window.mainloop()

def plot2():
    new_window = tk.Tk()
    new_window.title("Plot 2")

    fig, ax = plt.subplots()

    p2_input = txtpoly2.get()

    p2_coeffs = [float(x) for x in p2_input.split()]
    p2 = np.poly1d(p2_coeffs)

    x = np.linspace(-5, 5, 100)
    y = p2(x)
    ax.plot(x, y)
    ax.grid(True)

    poly_label = tk.Label(new_window, text=str(p2))
    poly_label.pack()

    canvas = FigureCanvasTkAgg(fig, master=new_window)
    canvas.draw()

    canvas.get_tk_widget().pack()

    new_window.mainloop()


title=tk.Label(main,text="Koeffizienten eingeben; mit l√ºcken zwischen den Zahlen",font=15)

labelframe=tk.Frame(main)
buttonframe=tk.Frame(main)

lblpoly1=tk.Label(labelframe,text="Polynom 1:", font=15)
lblpoly1.grid(row=0,column=0)

lblx1 = tk.Label(labelframe, text="X1:", font=15)
lblx1.grid(row=0, column=2)

txtx1 = tk.Entry(labelframe, font=15)
txtx1.grid(row=0, column=3)

lblx2 = tk.Label(labelframe, text="X2:", font=15)
lblx2.grid(row=1, column=2)

txtx2 = tk.Entry(labelframe, font=15)
txtx2.grid(row=1, column=3)

lblpoly2=tk.Label(labelframe, text="Polynom 2:", font=15)
lblpoly2.grid(row=1,column=0)

txtpoly1=tk.Entry(labelframe, font=15)
txtpoly1.grid(row=0,column=1)

txtpoly2=tk.Entry(labelframe,font=15)
txtpoly2.grid(row=1,column=1)

lbl_funktionswert = tk.Label(labelframe, text="Funktionswert: ",font=15)
lbl_funktionswert.grid(row=2,column=0)

txt_funktionswert = tk.Entry(labelframe, font=15)
txt_funktionswert.grid(row=2, column=1)

btn_multiplikation = tk.Button(buttonframe, text="Multiplizieren",font=15, command=multiplikation)
btn_multiplikation.grid(row=0,column=0,sticky="NESW")

btn_division = tk.Button(buttonframe, text="Division", font=15, command=division)
btn_division.grid(row=0,column=1,sticky="NESW")

btn_nullstellen = tk.Button(buttonframe,text="Nullstellen",font=15,command=nullstellen)
btn_nullstellen.grid(row=0,column=2,sticky="NESW")

btn_funktionswert = tk.Button(buttonframe, text="Gerade ermitteln", font=15, command=gerade)
btn_funktionswert.grid(row=0,column=3,sticky="NESW")

btn_funktionswert = tk.Button(buttonframe, text="Funktionswert", font=15,command=funktionswert)
btn_funktionswert.grid(row=0,column=4,sticky="NESW")

btn_ableitung1 = tk.Button(buttonframe, text="1. Ableitung", font=15,command=ableitung1)
btn_ableitung1.grid(row=1,column=0,sticky="NESW")

btn_ableitung2 = tk.Button(buttonframe, text="2. Ableitung", font=15, command=ableitung2)
btn_ableitung2.grid(row=1,column=1,sticky="NESW")

btn_ableitung3 = tk.Button(buttonframe, text="3. Ableitung", font=15, command=ableitung3)
btn_ableitung3.grid(row=1,column=2,sticky="NESW")

btn_stammfunktion = tk.Button(buttonframe, text="Stammfunktion", font=15, command=stammfunktion)
btn_stammfunktion.grid(row=1, column=3,sticky="NESW")

btnvieta = tk.Button(buttonframe, text="Vieta -p", font=15, command=vietap)
btnvieta.grid(row=1, column=4,sticky="NESW")

btnvieta = tk.Button(buttonframe, text="Vieta q", font=15, command=vietaq)
btnvieta.grid(row=1, column=5,sticky="NESW")

btnplot1 = tk.Button(buttonframe, text="Plot P1", font=15,command=plot1)
btnplot1.grid(row=2,column=0,sticky="NESW")

btnplot2 = tk.Button(buttonframe, text="Plot P2", font=15,command=plot2)
btnplot2.grid(row=2,column=1,sticky="NESW")


lbl_resultp1 = tk.Label(main, text="Resultat P1", font=15)
lbl_resultp2 = tk.Label(main, text="Resultat P2", font=15)

title.pack()
labelframe.pack()
buttonframe.pack()
lbl_resultp1.pack(side="left",expand=True)
lbl_resultp2.pack(side="right",expand=True)

main.mainloop()

```
