```python
import tkinter as tk
import numpy as np

main = tk.Tk()
main.minsize(800,400)
main.title("Vektoren")

def betrag():
    try:
        A = np.array([float(txta1.get()) if txta1.get() else 0.0, float(txta2.get()) if txta2.get() else 0.0, float(txta3.get()) if txta3.get() else 0.0])
        B = np.array([float(txtb1.get()) if txtb1.get() else 0.0, float(txtb2.get()) if txtb2.get() else 0.0, float(txtb3.get()) if txtb3.get() else 0.0])
        C = np.array([float(txtc1.get()) if txtc1.get() else 0.0, float(txtc2.get()) if txtc2.get() else 0.0, float(txtc3.get()) if txtc3.get() else 0.0])
        A_B = round(np.sqrt(np.dot(A, A)),3)
        B_B = round(np.sqrt(np.dot(B, B)),3)
        C_B = round(np.sqrt(np.dot(C, C)),3)
        lblResult["text"] = "Der Betrag von Vektor A ist: "+str(A_B)+"\n"+"Der Betrag von Vektor B ist: "+str(B_B)+"\n"+"Der Betrag von Vektor C ist: "+str(C_B)
    except Exception as e:
        lblResult["text"] = "Betragfunktion funktioniert nicht: "

def winkel():
    try:
        A = np.array([float(txta1.get()) if txta1.get() else 0.0, float(txta2.get()) if txta2.get() else 0.0, float(txta3.get()) if txta3.get() else 0.0])
        B = np.array([float(txtb1.get()) if txtb1.get() else 0.0, float(txtb2.get()) if txtb2.get() else 0.0, float(txtb3.get()) if txtb3.get() else 0.0])
        betraga = np.sqrt(np.dot(A,A))
        betragb = np.sqrt(np.dot(B,B))
        cos_ab = np.dot(A,B)/(betraga*betragb)
        result = round(np.degrees(np.arccos(cos_ab)),3)
        lblResult["text"] = "Der Winkel von A und B ist:" + str(result) + "Â°"

    except:
        lblResult["text"] = "Winkel funktioniert nicht: " 

def skalarprodukt():
    try:
        A = np.array([float(txta1.get()) if txta1.get() else 0.0, float(txta2.get()) if txta2.get() else 0.0, float(txta3.get()) if txta3.get() else 0.0])
        B = np.array([float(txtb1.get()) if txtb1.get() else 0.0, float(txtb2.get()) if txtb2.get() else 0.0, float(txtb3.get()) if txtb3.get() else 0.0])
        result = np.dot(A,B)
        lblResult["text"] = "Skalarprodukt von A und B ist:" + str(result)
    except:
        lblResult["text"] = "Skalarprodukt funktioniert nicht: " 

def kreuzprodukt():
    try:
        A = np.array([float(txta1.get()) if txta1.get() else 0.0, float(txta2.get()) if txta2.get() else 0.0, float(txta3.get()) if txta3.get() else 0.0])
        B = np.array([float(txtb1.get()) if txtb1.get() else 0.0, float(txtb2.get()) if txtb2.get() else 0.0, float(txtb3.get()) if txtb3.get() else 0.0])
        result=np.cross(A,B)
        lblResult["text"]= "Kreuzprodukt von A und B ist:" +str(result)    
    except:
        lblResult["text"] = "Kreuzprodukt funktioniert nicht: " 

def spatprodukt():
    try:
        A = np.array([float(txta1.get()) if txta1.get() else 0.0, float(txta2.get()) if txta2.get() else 0.0, float(txta3.get()) if txta3.get() else 0.0])
        B = np.array([float(txtb1.get()) if txtb1.get() else 0.0, float(txtb2.get()) if txtb2.get() else 0.0, float(txtb3.get()) if txtb3.get() else 0.0])
        C = np.array([float(txtc1.get()) if txtc1.get() else 0.0, float(txtc2.get()) if txtc2.get() else 0.0, float(txtc3.get()) if txtc3.get() else 0.0])
        result = np.dot(C,np.cross(A,B))
        lblResult["text"] = "Spatprodukt von A, B und C ist:" + str(result) + " Raumeinheiten"
    except:
        lblResult["text"] = "Spatprodukt funktioniert nicht: "

def dyadischesprodukt():
    try:
        X = np.array([float(x) for x in txtx.get().split(',')])
        Y = np.array([float(y) for y in txty.get().split(',')])
        result = np.outer(X,Y)
        lblResult["text"] = str(result)
    except:
        lblResult["text"] = "Dyadisches Produkt funktioniert nicht: eingabe mit Komma?(,)"

frame = tk.Frame(main)
frame.columnconfigure(0, weight=1)
frame.columnconfigure(1,weight=1)
frame.columnconfigure(2,weight=1)
frame.columnconfigure(3,weight=1)
frame.columnconfigure(4,weight=1)

# Vektor A
lblA = tk.Label(frame, text="Vektor A:", font=15)
lblA.grid(row=0,column=0)

txta1 = tk.Entry(frame,width=3,font=15)
txta1.grid(row=0,column=1)
txta2 = tk.Entry(frame,width=3,font=15)
txta2.grid(row=0,column=2)
txta3 = tk.Entry(frame,width=3,font=15)
txta3.grid(row=0,column=3)

# Vektor B
lblB = tk.Label(frame, text="Vektor B:", font=15)
lblB.grid(row=1,column=0)

txtb1 = tk.Entry(frame,width=3,font=15)
txtb1.grid(row=1,column=1)
txtb2 = tk.Entry(frame,width=3,font=15)
txtb2.grid(row=1,column=2)
txtb3 = tk.Entry(frame,width=3,font=15)
txtb3.grid(row=1,column=3)

# Vektor C
lblC = tk.Label(frame, text="Vektor C:", font=15)
lblC.grid(row=2,column=0)

txtc1 = tk.Entry(frame,width=3,font=15)
txtc1.grid(row=2,column=1)
txtc2 = tk.Entry(frame,width=3,font=15)
txtc2.grid(row=2,column=2)
txtc3 = tk.Entry(frame,width=3,font=15)
txtc3.grid(row=2,column=3)

# X / Y Dyadisches Produkt
lblx = tk.Label(frame, text="X:", font=15)
lblx.grid(row=0,column=4)
txtx = tk.Entry(frame, font=15, width=10)
txtx.grid(row=0, column=5)
lbly = tk.Label(frame, text="Y:", font=15)
lbly.grid(row=1,column=4)
txty = tk.Entry(frame, font=15, width=10)
txty.grid(row=1,column=5)
lbl_eingabe = tk.Label(frame, text="Eingabe mit ','",font=15)
lbl_eingabe.grid(row=2, column=5)


# Buttons
btnbetrag = tk.Button(frame, text="Betrag A|B|C", font=15, command=betrag)
btnbetrag.grid(row=3,column=0)
btnwinkel = tk.Button(frame, text="Winkel A|B",font=15, command=winkel)
btnwinkel.grid(row=3,column=1)
btnskalarprodukt = tk.Button(frame, text="Skalarprodukt A|B",font=15, command=skalarprodukt)
btnskalarprodukt.grid(row=3,column=2)
btnkreuzprdukt = tk.Button(frame, text="Kreuzprodukt A|B",font=15,command=kreuzprodukt)
btnkreuzprdukt.grid(row=3,column=3)
btnspatprodukt = tk.Button(frame, text="Spatprodukt A|B|C",font=15, command=spatprodukt)
btnspatprodukt.grid(row=3,column=4)
btndyadischesprodukt = tk.Button(frame, text="Dyadisches Produkt X|Y", font=15,command=dyadischesprodukt)
btndyadischesprodukt.grid(row=3, column=5)

lblResult = tk.Label(main,text="Result", font=15)

frame.pack()
lblResult.pack()


main.mainloop()
```
